```python
import pandas as pd
from getSraRunsFromAccIds import runGetRunsFromSampleAcc
from copy import deepcopy
```

# Read sample table

```python
df_sample = pd.read_csv('../deidentified_data_tables/samples/tblASVsamples.csv', index_col=0)
df_sample.head()
```

# Get info of all SRA

```python
df_sra_info = None
for sample in df_sample.index:
    identifier = df_sample.loc[sample, 'Accession']
    res=runGetRunsFromSampleAcc(identifier)
    if res is not None:
        for row in res:
            curr_df = pd.DataFrame.from_dict(row, orient='index').T
            if df_sra_info is None:
                df_sra_info = deepcopy(curr_df)
            else:
                df_sra_info = pd.concat([df_sra_info, curr_df], ignore_index=True)
df_sra_info.to_csv('sra_info_stats.csv')
df_sra_info.head()
```

# Count how many samples that have no SRA?

```python
set(df_sample.Accession)-set(df_sra_info.Run)
```

```python
df_sra_stats = pd.read_csv('sra_info_stats.csv', index_col=0).rename({'Run':'Accession'}, axis=1)
df_sra_stats.head()
```

```python
df_sample_expanded = pd.merge(df_sample.reset_index(), df_sra_stats, left_on='Accession', right_on='Accession', how='left')
df_sample_expanded.head()
```

```python
df_sample_expanded.to_csv('tblASVsamples_expanded.csv')
```

# Get unique SRA

```python
df_joaotbl = pd.read_csv('tblFigshare2.csv', index_col=0)

bioprojects = ['PRJNA394877','PRJNA491657','PRJNA545312','PRJNA548153','PRJNA606262','PRJNA607574']
unique_sras = []
for sample in df_joaotbl.index:
    all_sras = []
    for x in list(df_joaotbl.loc[sample, bioprojects]):
        if str(x) != 'nan':
            if '&' in x:
                all_sras.append(x.split('&')[0])
                all_sras.append(x.split('&')[1])
            else:
                all_sras.append(x)
    if len(all_sras) == 0:
        print(sample, 'no sra found')
    elif len(all_sras) == 1:
        unique_sras.append([sample, all_sras[0]])
    else:
        df_sra_info = None
        for identifier in all_sras:
            res=runGetRunsFromSampleAcc(identifier)
            if res is not None:
                for row in res:
                    curr_df = pd.DataFrame.from_dict(row, orient='index').T
                    if df_sra_info is None:
                        df_sra_info = deepcopy(curr_df)
                    else:
                        df_sra_info = pd.concat([df_sra_info, curr_df], ignore_index=True)
        df_sra_info = df_sra_info.sort_values(by='ReleaseDate', ascending=False)
        unique_sras.append([sample, df_sra_info.iloc[0]['Run']])
df_unique_sras = pd.DataFrame(unique_sras, columns=['SampleID','Accession'])
df_unique_sras.head()
```

```python
df_sample = pd.merge(df_sample, df_unique_sras.set_index('SampleID'), left_index=True, right_index=True, how='left')
df_sample.head()
```

```python
df_sample.to_csv('../deidentified_data_tables/samples/tblASVsamples.csv')
```
